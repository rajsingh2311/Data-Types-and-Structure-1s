1. What are data structures, and why are they important
**Answer:- Data structures organize and manage data efficiently, enabling optimized access, modification, and storage, crucial for solving computational problems effectively.**

2.Explain the difference between mutable and immutable data types with examples

**Answer:- Mutable data types can be modified after creation, e.g., lists ([1, 2] → [1, 3]). Immutable types cannot be changed, e.g., tuples ((1, 2) remains (1, 2)).**

3. What are the main differences between lists and tuples in Python

**Answer:- Lists are mutable, allowing modifications, while tuples are immutable and cannot be changed. Lists use more memory but support more operations, whereas tuples are faster and suitable for fixed data storage.**

4.Describe how dictionaries store data?
**Answer:- Dictionaries store data as key-value pairs in Python. Keys act as unique identifiers, while values store associated data. They use hash tables for fast lookups, insertions, and updates, ensuring efficiency.**

5. Why might you use a set instead of a list in Python?
**Answer:- Sets are used over lists for their unique, unordered elements and faster membership testing. They eliminate duplicates automatically and support efficient mathematical operations like unions, intersections, and differences.**

6.What is a string in Python, and how is it different from a list

**Answer:- A string is an immutable sequence of characters, used for text. Unlike lists, strings cannot be modified, support fewer operations, and store only characters, while lists hold any data type.**

7. How do tuples ensure data integrity in Python

**Answer:- Tuples ensure data integrity by being immutable, preventing accidental or intentional modifications after creation. This stability makes them ideal for storing fixed, unchangeable data, ensuring consistency and reliability in programs.**

8.What is a hash table, and how does it relate to dictionaries in Python
**Answer:- A hash table is a data structure that stores key-value pairs, using a hash function to map keys to specific indices for fast lookups. In Python, dictionaries implement hash tables to enable efficient access, insertion, and deletion of data.**

9. Can lists contain different data types in Python

**Answer:- Yes, lists in Python can contain different data types. A single list can store integers, strings, floats, and even other lists or objects, making them versatile and flexible for handling diverse data in a single collection.**

10. Explain why strings are immutable in Python

** Answer:- Strings are immutable in Python to ensure data integrity and optimize performance. Once created, they cannot be modified, preventing accidental changes and enabling efficient memory usage, particularly in large-scale applications.**

11. What advantages do dictionaries offer over lists for certain tasks

**Answer:- Dictionaries offer faster lookups, insertions, and deletions with key-based access, making them ideal for tasks involving frequent searching or mapping data. Unlike lists, they do not require sequential searches.**

12.Describe a scenario where using a tuple would be preferable over a list **Answer:- Using a tuple is preferable when storing fixed, unchangeable data like coordinates or dates. Their immutability ensures data integrity and allows for faster access and memory efficiency compared to lists.**

13:- How do sets handle duplicate values in Python
**Answer:- Sets automatically remove duplicate values by only storing unique elements. When a duplicate is added, it is ignored, ensuring that each item in a set is distinct without requiring manual checks.**

14.How does the “in” keyword work differently for lists and dictionaries **Answer:- For lists, the "in" keyword checks if a value exists within the list, performing a linear search. For dictionaries, it checks if a key exists in the dictionary, offering faster lookups due to the hash table implementation.**
15:-Can you modify the elements of a tuple? Explain why or why not
**Answer:- No, you cannot modify the elements of a tuple because tuples are immutable in Python. Once created, their contents cannot be changed. This immutability ensures data integrity and allows tuples to be used as reliable, fixed collections**

16: What is a nested dictionary, and give an example of its use case
**Answer:-A nested dictionary is a dictionary where the values are themselves dictionaries. It allows representing complex hierarchical data. For example, storing student information where each student's details (name, age, grades) are stored as a dictionary inside the main dictionary.**

17: Describe the time complexity of accessing elements in a dictionary

**Answer:- Accessing elements in a dictionary has an average time complexity of O(1) due to the underlying hash table structure. This allows for constant-time lookups, making dictionary access much faster compared to other data structures like lists.**

18: In what situations are lists preferred over dictionaries
**Answer:- Lists are preferred over dictionaries when the data is ordered or needs to be accessed by index, such as when working with sequences or collections of items where key-value mapping is unnecessary. Lists are also simpler for iteration.**

19. Why are dictionaries considered unordered, and how does that affect data retrieval ** Answer:Dictionaries are unordered because their elements are stored in a hash table, without a defined order. This means the order of insertion doesn’t affect retrieval, focusing on key-based access for efficiency.**

20.Explain the difference between a list and a dictionary in terms of data retrieval.
 **Answer:- List vs Dictionary in Data Retrieval: Lists retrieve elements by index, requiring sequential searches, while dictionaries use keys for direct access, offering faster, constant-time retrieval through hashing, regardless of the order.**
